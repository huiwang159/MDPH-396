# Selecting Models to Tune

```python
import csv
import pandas
```


## Process Step

## Summary Statistics

```python
pandas.concat([ps_mlts, pt_mlts, cf_mlts, os_mlts]).groupby('Multioutput').mean().sort_values('Score')
# MultiOutputRegressor
```

```python
pandas.concat([ps_enss, pt_enss, cf_enss, os_enss]).groupby('Ensemble').mean().sort_values('Score')
# GradientBoostingRegressor, RandomForestRegressor, AdaBoostRegressor, ExtraTreesRegressor, BaggingRegressor
```

```python
pandas.concat([ps_bss, pt_bss, cf_bss, os_bss]).groupby('Base').mean().sort_values('Score')
# Ridge, LinearSVR, GaussianProcessRegressor, PassiveAggressiveRegressor, KNeighborsRegressor
# MLPRegressor, HuberRegressor, SGDRegressor, LassoLarsIC, ExtraTreeRegressor
# ElasticNet, DecisionTreeRegressor, Perceptron, OrthogonalMatchingPursuit
```
